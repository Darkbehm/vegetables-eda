### Ejercicio seis: muestre por category,item,variety que tanto % ha aumentado su precio

Ejemplo: si para las manzanas gala el primer precio que se registro fue 2.00 y el ultimo precio que se registro es 4.00 este aumento un 100%

```python
pd.DataFrame({
    "category": ["fruta","fruta"]
    ,"item" : ['Manzana','Manzana']
    , "variety" : ['Gala','egremont_russet']
    , "percentage-change-between-first-and-last-price" : [200,150]       
})
```

```python
asc = vdf.sort_values(by="date",ascending=True)
```

```python
desc = vdf.sort_values(by="date",ascending=False)
```

```python
first = asc.loc[:,["item","variety","date","price"]].groupby(["item","variety"], as_index = False).head(1)
```

```python
last = desc.loc[:,["item","variety","date","price"]].groupby(["item","variety"], as_index = False).head(1)
```

```python
first = first.drop(columns="date")
last = last.drop(columns="date")
```

```python
first = first.rename(columns={"price":"first_price"}).sort_values(by=["item"])
last = last.rename(columns={"price":"last_price"}).sort_values(by=["item"])
```

```python
result6 = pd.merge(first,last)
```

```python
result6["percentage-change-between-first-and-last-price"] = (result6["last_price"]-result6["first_price"])/result6["first_price"]
```

```python
result6["percentage-change-between-first-and-last-price"] = round(result6["percentage-change-between-first-and-last-price"]*100,2)
```

```python
result6= result6.drop(columns= [ "first_price","last_price" ])
```

```python
result6
```
